```python
import numpy as np
import pandas as pd
```

### 1. Data Cleaning and Preparation

```python
# Loading the dataset
data = pd.read_csv("cardiotocographic.csv")
```

```python
data
```

```python
# identifying missing values
data.isnull().sum()
```

```python
# Handling missing values through Imputation which means Replacing the null values with mean or median
data_imputed = data.fillna(data.median())
```

```python
data_imputed
```

```python
# verifying any null values is present after imputation or not
data_imputed.isnull().sum()
```

```python
# identifying any inconsistency in data tyoes
data.dtypes
```

#### All columns are of type float64, so there are no issues with data types.

```python
# Detecting outliers using the IQR method
Q1 = data_imputed.quantile(0.25)
Q3 = data_imputed.quantile(0.75)
IQR = Q3 - Q1

# Identifying outliers
outliers = ((data_imputed < (Q1 - 1.5 * IQR)) | (data_imputed > (Q3 + 1.5 * IQR)))

# Print the count of outliers for each column
print(outliers.sum())
```

### 2. Statistical Summary

```python
data.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot size and style
plt.figure(figsize=(16, 10))
sns.set(style="whitegrid")

# Generate histograms for each numerical variable
data_imputed.hist(bins=20, figsize=(18, 12), color='blue', edgecolor='black')

plt.suptitle('Distribution of Numerical Variables in the Cardiotocographic Dataset', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
# Set the plot size
plt.figure(figsize=(12, 10))

# Calculate the correlation matrix
corr_matrix = data_imputed.corr()

# Generate a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Set the title
plt.title('Correlation Heatmap of All Variables', fontsize=16)

# Display the heatmap
plt.show()
```

```python
# List of numerical columns to create violin plots
numerical_columns = ['LB', 'AC', 'FM', 'UC', 'DL', 'DS', 'DP', 'ASTV', 'MSTV', 'ALTV', 'MLTV']

# Loop through each numerical column and generate a violin plot
for column in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.violinplot(data=data_imputed, x='NSP', y=column, palette='Set2')
    plt.title(f'Violin Plot of {column} by NSP', fontsize=16)
    plt.xlabel('NSP Categories')
    plt.ylabel(column)
    plt.show()
```

### Key Observations:

#### LB, AC, UC, ASTV, Width, Tendency: These columns have fewer outliers, suggesting that these variables are relatively well-distributed but still contain some extreme cases.
